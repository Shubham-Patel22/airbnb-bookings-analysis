## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
airbnb_data.dropna(subset = ['name', 'host_name'], axis = 0, inplace=True)
````

````python
airbnb_data['minimum_nights'].describe()
````

````python
airbnb_data['rental_type'] = np.where(airbnb_data['minimum_nights'] >= 28, 'monthly', 'short-term')
````

````python
airbnb_data['review_categorization'] = np.where(airbnb_data['number_of_reviews'] == 0, 'no reviews', np.where(airbnb_data['number_of_reviews'] < 10, '1-9 reviews', np.where(airbnb_data['number_of_reviews'] < 100, '10-99 reviews', '100 or more reviews')))
````

````python
airbnb_data['host_type'] = np.where(airbnb_data['calculated_host_listings_count'] == 1, 'amateur', np.where(airbnb_data['calculated_host_listings_count'] < 6, 'seasoned', np.where(airbnb_data['calculated_host_listings_count'] < 10, 'professional', np.where(airbnb_data['calculated_host_listings_count'] <100, 'expert', 'master'))))
````

````python
airbnb_data['price'].value_counts().sort_index()
````

````python
airbnb_data.loc[airbnb_data['price'] == 0, 'availability_365'].value_counts()
````

````python
airbnb_data = airbnb_data.loc[airbnb_data['price'] != 0]
````

````python
airbnb_data.head()
````

````python
airbnb_data.shape
````

### What all manipulations have you done and insights you found?

I removed all rows where name and host_name is empty. These can be improperly collected data and removing this small number of rows doesn't affect the analysis. I have also removed the listings with price 0, since it doesn't make any sense for a listing to have price 0.

Since minimum_nights is the minimum number of days the property should be booked by the guests for the booking to be accepted, it should have sensible values according to the business. But here descriptive statistics, we see that it has a max value of 1250, which is too high to be sensible. But on seeing business of Airbnb, it becomes clear why it is so. Along with the short-term rentals for vacations, AIrbnb has long-term rentals called Monthly Rentals for extended periods of stay due to which we have such anomalies because the data for both the types of rentals is combined in the given data. So, any stay having minimum_nights < 28 is a short-term rental, while any stay having minimum_nights >= 28 is a monthly rental. To capture this distinction, I'm going to make a categorical variable.

I have made a categorical variable based on number_of_reviews. It has 4 categories: 0 reviews, 1-9 reviews(1 digit), 10-99(2 digit), 100<=(3 digit).

Based on calculated_host_listings_count, I have made a categorical variable denoting the type of host, with hosts having 1 listing classified as amateur, 2-5 listings as seasoned, 6-9 listings as professional, 10-99 listings as expert and 100 & more listings as master.
